```python


import tkinter as tk
from tkinter import ttk, messagebox
from deep_translator import GoogleTranslator
import pyttsx3



root = tk.Tk()
root.title("Language Translator")
root.geometry("700x500")
root.resizable(False, False)
root.configure(bg="#E8F4FF")


engine = pyttsx3.init()

voices = engine.getProperty("voices")
engine.setProperty("voice", voices[0].id)
engine.setProperty("rate", 150)


languages = {
    "English":"en",
    "Hindi":"hi",
    "Gujarati":"gu",
    "French":"fr",
    "German":"de",
    "Japanese":"ja",
    "Chinese":"zh-CN",
    "Spanish":"es",
    "Russian":"ru",
    "Arabic":"ar",
    "Korean":"ko",
    "Italian":"it",
    "Portuguese":"pt",
    "Turkish":"tr"
}


def translate():

    text = input_text.get("1.0", tk.END).strip()

    if text == "":
        messagebox.showwarning("Warning","Please Enter Text")
        return

    translated = GoogleTranslator(
        source=languages[source_lang.get()],
        target=languages[target_lang.get()]
    ).translate(text)

    output_text.delete("1.0", tk.END)
    output_text.insert(tk.END, translated)



def speak():
    print("Speak Button Clicked")   # Testing

    text = output_text.get("1.0", tk.END).strip()

    if text == "":
        print("No text found")
        return

    print(text)

    engine.say(text)
    engine.runAndWait()

    print("Speech Finished")


def speak():

    text = output_text.get("1.0", tk.END).strip()

    if text == "":
        messagebox.showwarning("Warning", "Nothing to Speak")
        return

    engine.stop()
    engine.say(text)
    engine.runAndWait()


def copy():

    text = output_text.get("1.0", tk.END)

    root.clipboard_clear()
    root.clipboard_append(text)

    messagebox.showinfo("Success", "Copied Successfully")


def clear():

    input_text.delete("1.0", tk.END)
    output_text.delete("1.0", tk.END)

def swap_language():

    first = source_lang.get()
    second = target_lang.get()

    source_lang.set(second)
    target_lang.set(first)
    

title = tk.Label(
    root,
    text="🌍 Language Translator",
    font=("Arial",22,"bold"),
    bg="#E8F4FF",
    fg="blue"
)

title.pack(pady=15)



tk.Label(
    root,
    text="Enter Text",
    bg="#E8F4FF",
    font=("Arial",12)
).pack()

input_text = tk.Text(
    root,
    width=60,
    height=6,
    font=("Arial",12)
)

input_text.pack(pady=10)



lang_frame = tk.Frame(root, bg="#E8F4FF")
lang_frame.pack(pady=10)

source_lang = ttk.Combobox(
    lang_frame,
    values=list(languages.keys()),
    state="readonly",
    width=18
)

source_lang.current(0)
source_lang.grid(row=0,column=0,padx=10)


target_lang = ttk.Combobox(
    lang_frame,
    values=list(languages.keys()),
    state="readonly",
    width=18
)

target_lang.current(1)
target_lang.grid(row=0,column=2,padx=10)



frame = tk.Frame(root,bg="#E8F4FF")
frame.pack(pady=15)

translate_btn = tk.Button(
    frame,
    text="Translate",
    command=translate,
    bg="#007BFF",
    fg="white",
    width=12
)

translate_btn.grid(row=0,column=0,padx=5)

speak_btn = tk.Button(
    frame,
    text="Speak",
    command=speak,
    bg="green",
    fg="white",
    width=12
)

speak_btn.grid(row=0,column=1,padx=5)

copy_btn = tk.Button(
    frame,
    text="Copy",
    command=copy,
    bg="orange",
    fg="white",
    width=12
)

copy_btn.grid(row=0,column=2,padx=5)

clear_btn = tk.Button(
    frame,
    text="Clear",
    command=clear,
    bg="red",
    fg="white",
    width=12
)

clear_btn.grid(row=0,column=3,padx=5)


tk.Label(
    root,
    text="Translated Text",
    bg="#E8F4FF",
    font=("Arial",12)
).pack()

output_text = tk.Text(
    root,
    width=60,
    height=6,
    font=("Arial",12)
)

output_text.pack(pady=10)



root.mainloop()
```
